```python
import warnings
warnings.filterwarnings('ignore')
import os

#libraries for handling data
import pandas as pd
import numpy as np

#libraries for data visualization
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
 
#libraries for modelling
from sklearn import datasets, linear_model
```

```python
#inputs
sFilePathA = 'C:/Users/reejung/Documents/GitHub/comparison test.xlsx'
sSheetNameA = 'Sheet1'

sFilePathB = 'C:/Users/reejung/Documents/GitHub/comparison test.xlsx'
sSheetNameB = 'Sheet2'

columnsToJoin = 'name', 'date' #can be single column
```

```python
oFileA = pd.ExcelFile(sFilePathA)
oFileB = pd.ExcelFile(sFilePathB)

oFileA = pd.read_excel(oFileA, sheet_name = sSheetNameA, header=0)
oFileB = pd.read_excel(oFileB, sheet_name= sSheetNameB, header=0)
```

```python
dfA = pd.DataFrame(oFileA)
dfA
```

```python
dfB = pd.DataFrame(oFileB)
dfB
```

```python
dfColumns = set(dfA.columns.tolist() + dfB.columns.tolist())

for i in dfColumns:
    if i not in dfB.columns:
        dfB[i] = "" #np.nan
    if i not in dfA.columns:
        dfA[i] = ""
```

```python
dfA, dfB
```

```python
df = (dfA.merge(dfB, how='outer', on=columnsToJoin, suffixes=['_A', '_B'])).fillna('')
df
```

```python
for i in dfColumns:
    print(i)
    if i not in columnsToJoin:
        df[i+"_Match?"] = np.where(df[i+'_A'] == df[i+'_B'], 'True', 'False')
```

```python
df
```

```python
df = df.reindex(sorted(df.columns), axis=1)
df
```

```python
cols = list(columnsToJoin) + df.columns.drop(list(columnsToJoin)).tolist()
cols
```

```python
df = df[cols]
df
```

```python
match_cols = [col for col in df.columns if 'Match?' in col]
```

```python
df['Match?'] = df[match_cols].apply(
    lambda x: ','.join(x), axis=1)
```

```python
df['Match?'] = np.where(df['Match?'].str.contains("False"), "F", "T")
```

```python
df = df.loc[df['Match?']=="F"]
```

```python
def color_False(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for
    strings containing False, black otherwise.
    """
    color = 'red' if val =='False' else 'black'
    
    return 'color: %s' % color 
```

```python
df
```

```python
df.style.applymap(color_False)
#df.data = df.data.drop('Number', axis=1)     
```

```python
writer = pd.ExcelWriter('Comparison.xlsx', engine='xlsxwriter')

df.to_excel(writer, sheet_name='Sheet1')

writer.save()
```
