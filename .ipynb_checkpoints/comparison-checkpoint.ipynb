```python
dfColumns = set(dfA.columns.tolist() + dfB.columns.tolist())

for i in dfColumns:
    if i not in dfB.columns:
        dfB[i] = "" #np.nan
    if i not in dfA.columns:
        dfA[i] = ""
```

```python
def highlight_max(x):
    return ['background-color: yellow' if v == x.max() else ''
                for v in x]
```

```python
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color
```

```python
def color_False(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for
    strings containing False, black otherwise.
    """
    color = 'red' if val =='False' else 'black'
    
    return 'color: %s' % color 
```
